## PARTE 2 ‚Äì M√©todos e Heran√ßa

```python
class HumorAluno_Sabriny:
    def __init__(self, data_Sabriny, humor_Sabriny, comentario_Sabriny):
        self.data_Sabriny = data_Sabriny
        self.humor_Sabriny = humor_Sabriny
        self.comentario_Sabriny = comentario_Sabriny

    def exibir_dados_Sabriny(self, indice=None):
        if indice:
            print(f"Registro #{indice}")
        print(f"Data: {self.data_Sabriny} | Humor: {self.humor_Sabriny} | Coment√°rio: {self.comentario_Sabriny}\n")

    def alterar_humor_Sabriny(self, novo_humor):
        self.humor_Sabriny = novo_humor
        print("Humor alterado com sucesso!")

    def atualizar_comentario_Sabriny(self, novo_comentario):
        self.comentario_Sabriny = novo_comentario
        print("Coment√°rio atualizado com sucesso!")


class HumorAlunoDetalhado_Sabriny(HumorAluno_Sabriny):
    def __init__(self, data_Sabriny, humor_Sabriny, comentario_Sabriny, nota_dia_Sabriny):
        super().__init__(data_Sabriny, humor_Sabriny, comentario_Sabriny)
        self.nota_dia_Sabriny = nota_dia_Sabriny  # novo atributo exclusivo

    def exibir_dados_completos_Sabriny(self, indice=None):
        if indice:
            print(f"Registro Detalhado #{indice}")
        self.exibir_dados_Sabriny()
        print(f"Nota do dia (0 a 10): {self.nota_dia_Sabriny}\n")

    def avaliar_dia_Sabriny(self):
        if self.nota_dia_Sabriny >= 7:
            print("Avalia√ß√£o: Dia excelente!")
        elif self.nota_dia_Sabriny >= 4:
            print("Avalia√ß√£o: Dia mediano.")
        else:
            print("Avalia√ß√£o: Dia dif√≠cil.")


def menu_Sabriny():
    registros = []

    while True:
        print("1. Cadastrar humor b√°sico")
        print("2. Cadastrar humor detalhado")
        print("3. Exibir registros")
        print("4. Alterar registro")
        print("5. Sair")
        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == '1':
            data = input("Data (AAAA-MM-DD): ")
            humor = input("Humor (feliz, triste, neutro): ").lower()
            comentario = input("Coment√°rio: ")
            registros.append(HumorAluno_Sabriny(data, humor, comentario))
            print("Registro b√°sico cadastrado!\n")

        elif escolha == '2':
            data = input("Data (AAAA-MM-DD): ")
            humor = input("Humor (feliz, triste, neutro): ").lower()
            comentario = input("Coment√°rio: ")
            try:
                nota = float(input("Nota do dia (0 a 10): "))
                if nota < 0 or nota > 10:
                    print("Nota inv√°lida. Deve ser entre 0 e 10.\n")
                    continue
            except ValueError:
                print("Entrada inv√°lida para nota.\n")
                continue

            registros.append(HumorAlunoDetalhado_Sabriny(data, humor, comentario, nota))
            print("Registro detalhado cadastrado!\n")

        elif escolha == '3':
            if not registros:
                print("Nenhum registro cadastrado.\n")
            else:
                for i, reg in enumerate(registros, 1):
                    if isinstance(reg, HumorAlunoDetalhado_Sabriny):
                        reg.exibir_dados_completos_Sabriny(i)
                        reg.avaliar_dia_Sabriny()
                    else:
                        reg.exibir_dados_Sabriny(i)

        elif escolha == '4':
            if not registros:
                print("Nenhum registro para alterar.\n")
                continue

            for i, reg in enumerate(registros, 1):
                tipo = "Detalhado" if isinstance(reg, HumorAlunoDetalhado_Sabriny) else "B√°sico"
                print(f"{i}. {reg.data_Sabriny} - {reg.humor_Sabriny} ({tipo})")

            try:
                num = int(input("N√∫mero do registro para alterar: "))
                if num < 1 or num > len(registros):
                    print("N√∫mero inv√°lido.\n")
                    continue
            except ValueError:
                print("Entrada inv√°lida.\n")
                continue

            reg = registros[num - 1]

            novo_humor = input("Novo humor (deixe vazio para n√£o alterar): ").lower()
            if novo_humor:
                reg.alterar_humor_Sabriny(novo_humor)

            novo_comentario = input("Novo coment√°rio (deixe vazio para n√£o alterar): ")
            if novo_comentario:
                reg.atualizar_comentario_Sabriny(novo_comentario)

            if isinstance(reg, HumorAlunoDetalhado_Sabriny):
                try:
                    nova_nota = input("Nova nota do dia (0 a 10, vazio para n√£o alterar): ")
                    if nova_nota:
                        nova_nota = float(nova_nota)
                        if 0 <= nova_nota <= 10:
                            reg.nota_dia_Sabriny = nova_nota
                            print("Nota atualizada com sucesso!")
                        else:
                            print("Nota inv√°lida. Ignorando altera√ß√£o.")
                except ValueError:
                    print("Entrada inv√°lida para nota. Ignorando altera√ß√£o.")

            print()

        elif escolha == '5':
            print("Saindo do sistema. At√© mais!")
            break

        else:
            print("Op√ß√£o inv√°lida.\n")


menu_Sabriny()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
class HumorAluno_Sabriny:
    def __init__(self, data_Sabriny, humor_Sabriny, comentario_Sabriny):
        self.data_Sabriny = data_Sabriny
        self.humor_Sabriny = humor_Sabriny
        self.__comentario_Sabriny = comentario_Sabriny  # atributo privado

    # Getter para coment√°rio
    def get_comentario_Sabriny(self):
        return self.__comentario_Sabriny

    # Setter para coment√°rio
    def set_comentario_Sabriny(self, novo_comentario):
        if isinstance(novo_comentario, str) and len(novo_comentario) <= 200:
            self.__comentario_Sabriny = novo_comentario
            print("Coment√°rio atualizado com sucesso!")
        else:
            print("Coment√°rio inv√°lido ou muito longo (m√°x 200 caracteres).")

    def exibir_dados_Sabriny(self, indice=None):
        if indice:
            print(f"Registro #{indice}")
        print(f"Data: {self.data_Sabriny} | Humor: {self.humor_Sabriny} | Coment√°rio: {self.__comentario_Sabriny}\n")

    # M√©todo com l√≥gica: verifica se humor √© positivo
    def humor_positivo_Sabriny(self):
        return self.humor_Sabriny.lower() == 'feliz'


def menu_Sabriny():
    registros = []

    while True:
        print("1. Cadastrar humor")
        print("2. Exibir registros")
        print("3. Atualizar coment√°rio")
        print("4. Verificar se humor √© positivo")
        print("5. Sair")
        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == '1':
            data = input("Data (AAAA-MM-DD): ")
            humor = input("Humor (feliz, triste, neutro): ").lower()
            comentario = input("Coment√°rio: ")
            registros.append(HumorAluno_Sabriny(data, humor, comentario))
            print("Registro cadastrado!\n")

        elif escolha == '2':
            if not registros:
                print("Nenhum registro cadastrado.\n")
            else:
                for i, reg in enumerate(registros, 1):
                    reg.exibir_dados_Sabriny(i)

        elif escolha == '3':
            if not registros:
                print("Nenhum registro para atualizar.\n")
                continue

            for i, reg in enumerate(registros, 1):
                print(f"{i}. Data: {reg.data_Sabriny} - Humor: {reg.humor_Sabriny}")

            try:
                num = int(input("N√∫mero do registro para atualizar coment√°rio: "))
                if 1 <= num <= len(registros):
                    novo_com = input("Digite o novo coment√°rio (m√°x 200 caracteres): ")
                    registros[num-1].set_comentario_Sabriny(novo_com)
                else:
                    print("N√∫mero inv√°lido.\n")
            except ValueError:
                print("Entrada inv√°lida.\n")

        elif escolha == '4':
            if not registros:
                print("Nenhum registro para verificar.\n")
                continue

            for i, reg in enumerate(registros, 1):
                positivo = reg.humor_positivo_Sabriny()
                status = "positivo" if positivo else "n√£o positivo"
                print(f"Registro #{i} em {reg.data_Sabriny}: Humor √© {status}.")

            print()

        elif escolha == '5':
            print("Saindo do sistema. At√© mais!")
            break

        else:
            print("Op√ß√£o inv√°lida.\n")


menu_Sabriny()
```

```python
import sqlite3

#Classe principal
class HumorAluno_Sabriny:
    def __init__(self, data_Sabriny, humor_Sabriny, comentario_Sabriny):
        self.data_Sabriny = data_Sabriny
        self.humor_Sabriny = humor_Sabriny
        self.comentario_Sabriny = comentario_Sabriny

    def exibir_dados_Sabriny(self):
        print(f"Data: {self.data_Sabriny} | Humor: {self.humor_Sabriny} | Coment√°rio: {self.comentario_Sabriny}")

    def salvar_no_bd_Sabriny(self, cursor):
        cursor.execute(
            "INSERT INTO humor (data, humor, comentario, tipo) VALUES (?, ?, ?, ?)",
            (self.data_Sabriny, self.humor_Sabriny, self.comentario_Sabriny, "basico")
        )


# Subclasse com polimorfismo
class HumorAlunoDetalhado_Sabriny(HumorAluno_Sabriny):
    def __init__(self, data_Sabriny, humor_Sabriny, comentario_Sabriny, nota_dia_Sabriny):
        super().__init__(data_Sabriny, humor_Sabriny, comentario_Sabriny)
        self.nota_dia_Sabriny = nota_dia_Sabriny

    # Sobrescreve m√©todo para exibir dados completos
    def exibir_dados_Sabriny(self):
        super().exibir_dados_Sabriny()
        print(f"Nota do dia: {self.nota_dia_Sabriny}")

    def salvar_no_bd_Sabriny(self, cursor):
        cursor.execute(
            "INSERT INTO humor (data, humor, comentario, tipo, nota) VALUES (?, ?, ?, ?, ?)",
            (self.data_Sabriny, self.humor_Sabriny, self.comentario_Sabriny, "detalhado", self.nota_dia_Sabriny)
        )


# Fun√ß√£o para criar tabela SQLite (se n√£o existir)
def criar_tabela_Sabriny(con):
    cursor = con.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS humor (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            data TEXT NOT NULL,
            humor TEXT NOT NULL,
            comentario TEXT,
            tipo TEXT NOT NULL,
            nota REAL
        )
    ''')
    con.commit()


def menu_Sabriny():
    con = sqlite3.connect('humor_alunos_sabriny.db')
    criar_tabela_Sabriny(con)
    cursor = con.cursor()

    while True:
        print("\n1. Cadastrar humor b√°sico")
        print("2. Cadastrar humor detalhado")
        print("3. Exibir registros")
        print("4. Excluir registro")
        print("5. Sair")
        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == '1':
            data = input("Data (AAAA-MM-DD): ")
            humor = input("Humor (feliz, triste, neutro): ").lower()
            comentario = input("Coment√°rio: ")
            reg = HumorAluno_Sabriny(data, humor, comentario)
            reg.salvar_no_bd_Sabriny(cursor)
            con.commit()
            print("Registro b√°sico salvo no banco.\n")

        elif escolha == '2':
            data = input("Data (AAAA-MM-DD): ")
            humor = input("Humor (feliz, triste, neutro): ").lower()
            comentario = input("Coment√°rio: ")
            try:
                nota = float(input("Nota do dia (0 a 10): "))
                if nota < 0 or nota > 10:
                    print("Nota inv√°lida. Deve ser entre 0 e 10.\n")
                    continue
            except ValueError:
                print("Entrada inv√°lida para nota.\n")
                continue
            reg = HumorAlunoDetalhado_Sabriny(data, humor, comentario, nota)
            reg.salvar_no_bd_Sabriny(cursor)
            con.commit()
            print("Registro detalhado salvo no banco.\n")

        elif escolha == '3':
            cursor.execute("SELECT id, data, humor, comentario, tipo, nota FROM humor")
            rows = cursor.fetchall()
            if not rows:
                print("Nenhum registro encontrado.\n")
            else:
                for row in rows:
                    id_, data, humor, comentario, tipo, nota = row
                    print(f"ID: {id_} | Data: {data} | Humor: {humor} | Coment√°rio: {comentario} | Tipo: {tipo}", end="")
                    if tipo == "detalhado":
                        print(f" | Nota: {nota}")
                    else:
                        print()

        elif escolha == '4':
            try:
                id_del = int(input("ID do registro para excluir: "))
                cursor.execute("DELETE FROM humor WHERE id = ?", (id_del,))
                if cursor.rowcount > 0:
                    con.commit()
                    print("Registro exclu√≠do.\n")
                else:
                    print("ID n√£o encontrado.\n")
            except ValueError:
                print("Entrada inv√°lida.\n")

        elif escolha == '5':
            print("Saindo do sistema...")
            break

        else:
            print("Op√ß√£o inv√°lida.\n")

    con.close()

menu_Sabriny()
```

```python
import mysql.connector

# === CLASSE PRINCIPAL - HumorAluno_Sabriny ===
class HumorAluno_Sabriny:
    def __init__(self, data_Sabriny, humor_Sabriny, comentario_Sabriny):
        self.data_Sabriny = data_Sabriny
        self.humor_Sabriny = humor_Sabriny
        self.__comentario_Sabriny = comentario_Sabriny

    def get_comentario_Sabriny(self):
        return self.__comentario_Sabriny

    def set_comentario_Sabriny(self, novo_comentario):
        if isinstance(novo_comentario, str) and len(novo_comentario) <= 200:
            self.__comentario_Sabriny = novo_comentario
            print("Coment√°rio atualizado com sucesso!")
        else:
            print("Coment√°rio inv√°lido ou muito longo (m√°x 200 caracteres).")

    def exibir_dados_Sabriny(self):
        print(f"Data: {self.data_Sabriny} | Humor: {self.humor_Sabriny} | Coment√°rio: {self.__comentario_Sabriny}")

    def humor_positivo_Sabriny(self):
        return self.humor_Sabriny.lower() == 'feliz'

    def salvar_no_bd_Sabriny(self, cursor):
        sql = "INSERT INTO humor (data, humor, comentario, tipo) VALUES (%s, %s, %s, %s)"
        valores = (self.data_Sabriny, self.humor_Sabriny, self.__comentario_Sabriny, "basico")
        cursor.execute(sql, valores)


# === SUBCLASSE - HumorAlunoDetalhado_Sabriny ===
class HumorAlunoDetalhado_Sabriny(HumorAluno_Sabriny):
    def __init__(self, data_Sabriny, humor_Sabriny, comentario_Sabriny, nota_dia_Sabriny):
        super().__init__(data_Sabriny, humor_Sabriny, comentario_Sabriny)
        self.nota_dia_Sabriny = nota_dia_Sabriny

    def exibir_dados_Sabriny(self):
        super().exibir_dados_Sabriny()
        print(f"Nota do dia: {self.nota_dia_Sabriny}")

    def salvar_no_bd_Sabriny(self, cursor):
        sql = "INSERT INTO humor (data, humor, comentario, tipo, nota) VALUES (%s, %s, %s, %s, %s)"
        valores = (self.data_Sabriny, self.humor_Sabriny, self.get_comentario_Sabriny(), "detalhado", self.nota_dia_Sabriny)
        cursor.execute(sql, valores)


# === FUN√á√ÉO PRINCIPAL COM MENU INTERATIVO ===
def menu_Sabriny():
    try:

        
        con = mysql.connector.connect(
            host="localhost",
            user="root",
            password="",
            database="humor_alunos_sabriny"
        )
        cursor = con.cursor()




    
    except mysql.connector.Error as err:
        print(f"Erro ao conectar ao banco de dados: {err}")
        return

    while True:
        print("\n===== Sistema de Registro de Humor Di√°rio - Sabriny =====")
        print("1. Cadastrar humor b√°sico")
        print("2. Cadastrar humor detalhado")
        print("3. Exibir todos os registros")
        print("4. Atualizar coment√°rio de um registro")
        print("5. Excluir registro")
        print("6. Verificar quais humores s√£o positivos")
        print("7. Sair")
        escolha = input("Escolha uma op√ß√£o: ")

        if escolha == '1':
            data = input("Data (AAAA-MM-DD): ")
            humor = input("Humor (feliz, triste, neutro): ").lower()
            comentario = input("Coment√°rio: ")
            reg = HumorAluno_Sabriny(data, humor, comentario)
            reg.salvar_no_bd_Sabriny(cursor)
            con.commit()
            print("Registro b√°sico salvo com sucesso!\n")

        elif escolha == '2':
            data = input("Data (AAAA-MM-DD): ")
            humor = input("Humor (feliz, triste, neutro): ").lower()
            comentario = input("Coment√°rio: ")
            try:
                nota = float(input("Nota do dia (0 a 10): "))
                if not (0 <= nota <= 10):
                    print("Nota inv√°lida, deve ser entre 0 e 10.\n")
                    continue
            except ValueError:
                print("Entrada inv√°lida para nota.\n")
                continue
            reg = HumorAlunoDetalhado_Sabriny(data, humor, comentario, nota)
            reg.salvar_no_bd_Sabriny(cursor)
            con.commit()
            print("Registro detalhado salvo com sucesso!\n")

        elif escolha == '3':
            cursor.execute("SELECT id, data, humor, comentario, tipo, nota FROM humor")
            rows = cursor.fetchall()
            if not rows:
                print("Nenhum registro encontrado.\n")
            else:
                for row in rows:
                    id_, data, humor, comentario, tipo, nota = row
                    print(f"ID: {id_} | Data: {data} | Humor: {humor} | Coment√°rio: {comentario} | Tipo: {tipo}", end="")
                    if tipo == "detalhado":
                        print(f" | Nota: {nota}")
                    else:
                        print()

        elif escolha == '4':
            try:
                id_upd = int(input("Digite o ID do registro para atualizar coment√°rio: "))
                novo_com = input("Digite o novo coment√°rio (m√°x 200 caracteres): ")
                if len(novo_com) > 200:
                    print("Coment√°rio muito longo.\n")
                    continue
                cursor.execute("UPDATE humor SET comentario = %s WHERE id = %s", (novo_com, id_upd))
                if cursor.rowcount == 0:
                    print("ID n√£o encontrado.\n")
                else:
                    con.commit()
                    print("Coment√°rio atualizado com sucesso!\n")
            except ValueError:
                print("Entrada inv√°lida.\n")

        elif escolha == '5':
            try:
                id_del = int(input("Digite o ID do registro para excluir: "))
                cursor.execute("DELETE FROM humor WHERE id = %s", (id_del,))
                if cursor.rowcount == 0:
                    print("ID n√£o encontrado.\n")
                else:
                    con.commit()
                    print("Registro exclu√≠do com sucesso!\n")
            except ValueError:
                print("Entrada inv√°lida.\n")

        elif escolha == '6':
            cursor.execute("SELECT id, data, humor FROM humor")
            rows = cursor.fetchall()
            if not rows:
                print("Nenhum registro para verificar.\n")
            else:
                print("Registros com humor positivo (feliz):")
                for id_, data, humor in rows:
                    if humor.lower() == 'feliz':
                        print(f"ID: {id_} | Data: {data} | Humor: {humor}")
                print()

        elif escolha == '7':
            print("Saindo do sistema. At√© mais!")
            break

        else:
            print("Op√ß√£o inv√°lida. Tente novamente.\n")

    cursor.close()
    con.close()


# === EXECUTA O MENU PRINCIPAL ===
if __name__ == "__main__":
    menu_Sabriny()
```
